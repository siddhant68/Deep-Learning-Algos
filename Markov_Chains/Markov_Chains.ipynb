```python
T = {}

order = 4

for ix in range(len(data)-order):
    # current context
    ctx = data[ix:ix+order]
    
    #get future value
    next_letter = data[ix+order]
    
    #check if context exists in dict
    if T.get(ctx) is None:
        T[ctx] = {}
        T[ctx][next_letter] = 1
    else:
        if T[ctx].get(next_letter) is None:
            T[ctx][next_letter] = 1
        else:
            T[ctx][next_letter] += 1
```

```python
T
```

```python
for kx in T.keys():
    s = float(sum(T[kx].values()))
    
    for k in T[kx].keys():
        T[kx][k] = T[kx][k]/s
```

```python
def generate_next(ctx):
    
    possible = T.get(ctx)
    
    if possible is None:
        return " "
    
    shuffled_keys = possible.keys()
    random.shuffle(shuffled_keys)
    
    keys_probs = [possible[kx] for kx in shuffled_keys]
    
    return np.random.choice(shuffled_keys,p=keys_probs)
```
